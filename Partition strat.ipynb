```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(0,1, 25000)
data2 = np.random.normal(0,1, 25000)

data = np.append(data, np.random.normal(-10,1, 25000))
data2 = np.append(data2, np.random.normal(-10,1, 25000))

data_set = np.column_stack((data,data2))
np.savetxt('input.csv',data_set,delimiter=',')
```

```python
plt.scatter(data_set[:,0],data_set[:,1])
```

```python
scores = np.loadtxt('out.csv')
```

```python
alpha = .515
mapping = []
for i in scores:
    if(i < alpha):
        mapping.append(1)
    else:
        mapping.append(-1)
```

```python
inlier = []
outlier = []
for count,i in enumerate(mapping):
    if(mapping[count] == 1):
        inlier.append(data_set[count])
    else:
        outlier.append( data_set[count])
inlier = np.array(inlier)
outlier = np.array(outlier)

plt.scatter(outlier[:,0],outlier[:,1], c='black',  s = .5)
plt.scatter(inlier[:,0],inlier[:,1], c='red', s = .5)
```
