```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cov1 = np.matrix([[1,0],[0,1]])
cov2 = np.matrix([[1,0],[0,1]])

mean1 = np.array([0,0])
mean2 = np.array([-7,-7])
                 
dist1 = np.random.multivariate_normal(mean1, cov1, size=15000)
dist2 = np.random.multivariate_normal(mean2, cov2, size=20000)

data_set = np.append(dist1,dist2, axis=0)
np.savetxt('input.csv',data_set,delimiter=',')
```

```python
plt.scatter(data_set[:,0],data_set[:,1])
```

```python
os.system("python main.py input.csv out.csv")
```

```python
scores = np.loadtxt('out.csv')
```

```python
alpha = .563
mapping = []
for i in scores:
    if(i < alpha):
        mapping.append(1)
    else:
        mapping.append(-1)
```

```python
inlier = []
outlier = []
for count,i in enumerate(mapping):
    if(mapping[count] == 1):
        inlier.append(data_set[count])
    else:
        outlier.append( data_set[count])
inlier = np.array(inlier)
outlier = np.array(outlier)

plt.scatter(outlier[:,0],outlier[:,1], c='red',  s = .5, label="Anomalous")
plt.scatter(inlier[:,0],inlier[:,1], c='black', s = .5, label= "Nominal")
plt.legend(loc="upper left")
```
